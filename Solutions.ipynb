# **PROBLEM 1:**

**PART 1:** *This function solves the Greatest Common Divisor (GCD) of two integers using the Euclidean Algorithm.
GCD is the largest number that divides by both the numbers without leaving a remainder.*

````python
def gcd(m, n):
    while n != 0:
        m, n = n, m % n
    return m

print(gcd(16, 4))
print(gcd(75, 25))
print(gcd(21, 8))
print(gcd(69, 33))
````

**PART 2:** *What the function does if one or both of the arguments are negative*

The GCD is always a positive integer, so its right. The modulo operation in Python also handles negatives as Python's % always returns a non-negative result when the divisor is +ve.

````python
print(gcd(-69, 33))
print(gcd(69, -33))
print(gcd(-69, -33))
````

# **PROBLEM 2:**
**PART 1:** 

````python
import math

def euler_limit(n):
    return (1 + 1/n) ** n

print(euler_limit(10))
print(euler_limit(100))
print(euler_limit(1000))
print(euler_limit(1000000))
print("True e =", math.exp(1))
````

**PART 2:**

If euler_limit(n) is called with a huge value like 10^{21}, the function will return 1.0.
it will round off "1/n" to zero and the base will become 1.0 and 1.0 raised to any power will be 1.

````python
print(euler_limit(10**10))
print(euler_limit(10**11))
print(euler_limit(10**19))
print(euler_limit(10**31))
````

**PART 3:**




````python
import math
def factorial(k):
    if k == 0:
        return 1
    return k * factorial(k - 1)

def euler_sum(n):
    total = 0.0
    for k in range(n):
        total += 1 / factorial(k)
    return total

print(euler_sum(0))
print(euler_sum(6))
print(euler_sum(8))
print(euler_sum(10))
print(euler_sum(40))
print("True e =", math.exp(1))
````

**PART 4:**


````python
def euler_approx(epsilon):
    true_e = math.exp(1)
    n = 1
    while True:
        approx = euler_sum(n)
        if abs(approx - true_e) < epsilon:
            return approx
        n += 1

print(euler_approx(0.1))
print(euler_approx(0.01))
print(euler_approx(0.0001))
print(euler_approx(0.000001))
````

**PART 5:**


````python
def print_euler_sum_error(n):
    true_e = math.exp(1)
    for k in range(1, n + 1):
        error = euler_sum(k) - true_e
        print(error)

def print_euler_lim_error(n):
    true_e = math.exp(1)
    for k in range(1, n + 1):
        error = euler_limit(k) - true_e
        print(error)

print("euler_sum error")
print_euler_sum_error(4)

print("\n euler_limit error")
print_euler_lim_error(4)
````

**PROBLEM 6:**

the euler_sum approximation is better. It's faster and more precise than the limit approach.
